```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
balls = cv2.imread("images.jpg")
gray_img = cv2.cvtColor(balls, cv2.COLOR_BGR2GRAY)
gray_img = cv2.medianBlur(gray_img, 5)
```

```python
circles = cv2.HoughCircles(gray_img,cv2.HOUGH_GRADIENT,1,50,
 param1=200,param2=20,minRadius=0,maxRadius=0)
```

```python
circles = np.uint16(np.around(circles))

for i in circles[0,:]:
 # draw the outer circle
 cv2.circle(balls,(i[0],i[1]),i[2],(0,0,255),2)
```

```python
cv2.imwrite("planets_circles.jpg", balls)
cv2.imshow("HoughCirlces", balls)
cv2.waitKey()
cv2.destroyAllWindows()
```
